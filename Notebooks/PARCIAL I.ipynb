```python
import pandas as pd
import numpy as np 
```

```python

emicron = pd.read_csv("https://github.com/niconomist98/DataAnalyticsUQ/raw/refs/heads/main/Datos/EMICRON/emicron_identificacion.csv")
```

```python
emicron
```

```python
#micronegicios en el quindio cuyos propietarios son hombres entre 18 y 30 aÃ±os 
df_d_genero = emicron[emicron['P3271'].isin([1,2])]





microneg_quindio = emicron.query((df_d_genero['P3271'] == 1)&"COD_DEPTO == 63 and 18 <= P241 <= 30")['F_EXP'].sum()

print(f"Cantidad de micronegocios en el QuindÃ­o encabezados por personas con edades entre 18 y 30 aÃ±os: {int(microneg_quindio)}")
```

```python
#micronegocios de la industra manufacturera existen en risaralda cutas propietarias son mujeres entre 20 y 30 aÃ±os  
# P35: gÃ©nero (1 = hombre, 2 = mujer), P241: edad, COD_DEPTO: departamento (66 = Risaralda), GRUPOS4: industria manufacturera (10-33)

microneg_risaralda = emicron.query("P35 == 2 and COD_DEPTO == 66 and 20 <= P241 <= 30 & GRUPOS4 == 2")['F_EXP'].sum()
print(f"Cantidad de micronegocios de la industria manufacturera en Risaralda encabezados por mujeres entre 20 y 30 aÃ±os: {int(microneg_risaralda)}")

```

```python
#ENTRE MANIZALES ARMENIA Y PEREIRA CUAL TIENE MAS MICRONEGOCIOS DEL SECTOR COMERCIO CUYAS PROPIERTARIAS SON MUJERES MAYORES DE 40 AÃ‘OS
# P35: gÃ©nero (1 = hombre, 2 = mujer), P241: edad, COD_DEPTO: departamento (17 = Caldas, 63 = QuindÃ­o, 66 = Risaralda), GRUPOS4: comercio (45-47)

# micronegocios del sector comercio en el QuindÃ­o cuyas propietarias son mujeres mayores de 40 aÃ±os
microneg_comercio_quindio = emicron.query("(P35 == 2) & (COD_DEPTO == 63) & (P241 > 40) & (GRUPOS4 == 3)")['F_EXP'].sum()
print(f"Cantidad de micronegocios del sector comercio en el QuindÃ­o encabezados por mujeres mayores de 40 aÃ±os: {int(microneg_comercio_quindio)}")

# micronegocios del sector comercio en Risaralda cuyas propietarias son mujeres mayores de 40 aÃ±os
microneg_comercio_risaralda = emicron.query("(P35 == 2) & (COD_DEPTO == 66) & (P241 > 40) & (GRUPOS4 == 3)")['F_EXP'].sum()
print(f"Cantidad de micronegocios del sector comercio en el Risaralda encabezados por mujeres mayores de 40 aÃ±os: {int(microneg_comercio_risaralda)}")

# micronegocios del sector comercio en Caldas cuyas propietarias son mujeres mayores de 40 aÃ±os
microneg_comercio_caldas = emicron.query("(P35 == 2) & (COD_DEPTO == 17) & (P241 > 40) & (GRUPOS4 == 3)")['F_EXP'].sum()
print(f"Cantidad de micronegocios del sector comercio en el Caldas encabezados por mujeres mayores de 40 aÃ±os: {int(microneg_comercio_caldas)}")
```

```python
#En la cuidad de armenia cunatos micronegocios cuyos propietarios son hombres no to tienen nombre comercial 
# P35: gÃ©nero (1 = hombre, 2 = mujer), NOMBRE_COMERCIAL: P3035 DEPTO: departamento (63 = QuindÃ­o)

microneg_sin_nombre = emicron.query("P35 == 1 & COD_DEPTO == 63 & P3035 == 2")['F_EXP'].sum()
print(f"Cantidad de micronegocios en el QuindÃ­o encabezados por hombres sin nombre comercial: {int(microneg_sin_nombre)}")
```

```python
#En las ciudades de Armenia y Pereira, Â¿cuÃ¡l tiene mÃ¡s micronegocios cuyos propietarios son hombres mayores de 20 aÃ±os? 
# P35: gÃ©nero (1 = hombre, 2 = mujer), P241: edad, COD_DEPTO: departamento (63 = QuindÃ­o, 66 = Risaralda)
microneg_hombres_mayores_20 = emicron.query("(P35 == 1) & (P241 > 20) & (COD_DEPTO in [63, 66])").groupby('COD_DEPTO')['F_EXP'].sum()
ciudades = {63: 'Armenia', 66: 'Pereira'}
ciudad_mas_micronegocios = microneg_hombres_mayores_20.idxmax()
cantidad_mas_micronegocios = microneg_hombres_mayores_20.max()
print(f"La ciudad con mÃ¡s micronegocios cuyos propietarios son hombres mayores de 20 aÃ±os es {ciudades[ciudad_mas_micronegocios]} con {int(cantidad_mas_micronegocios)} micronegocios.")        
```
